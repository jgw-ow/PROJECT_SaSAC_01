```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 파일 경로와 연도를 인자로 받아 합계 데이터 추출
def extract_relevant_totals(file_path, year):
    df = pd.ExcelFile(file_path).parse('Main View')
    df = df.loc[:, ~df.columns.duplicated()]  # 중복 열 제거
    df_totals = df[['합계', 'Unnamed: 8']].iloc[1]  # 합계 데이터 추출
    return pd.DataFrame({'년': [year], '발생량': [df_totals['합계']], '도착량': [df_totals['Unnamed: 8']]})

# 연도별 데이터 처리
data_2020 = extract_relevant_totals('성수동1,2가_2020.xlsx', 2020)
data_2021 = extract_relevant_totals('성수동1,2가_2021.xlsx', 2021)
data_2022 = extract_relevant_totals('성수동1,2가_2022.xlsx', 2022)
data_2023 = extract_relevant_totals('성수동1,2가_2023.xlsx', 2023)

# 데이터 통합 및 추가 계산
all_yearly_totals = pd.concat([data_2020, data_2021, data_2022, data_2023], ignore_index=True)
all_yearly_totals['발생량'] = pd.to_numeric(all_yearly_totals['발생량'], errors='coerce')
all_yearly_totals['도착량'] = pd.to_numeric(all_yearly_totals['도착량'], errors='coerce')

# 발생량 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(all_yearly_totals['년'], all_yearly_totals['발생량'], marker='o', label='발생량')
plt.title('성수동 1,2가 연도별 발생량 (2020년 ~ 2023년)')
plt.xlabel('년도')
plt.ylabel('발생량 (천만)')
plt.xticks(range(2020, 2024))  # X축 범위 설정
plt.grid()
plt.legend()
plt.show()

# 도착량 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(all_yearly_totals['년'], all_yearly_totals['도착량'], marker='o', label='도착량')
plt.title(' 성수동 1,2가 연도별 도착량 (2020년 ~ 2023년)')
plt.xlabel('년도')
plt.ylabel('도착량 (천만)')
plt.xticks(range(2020, 2024))  # X축 범위 설정
plt.grid()
plt.legend()
plt.show()
```
